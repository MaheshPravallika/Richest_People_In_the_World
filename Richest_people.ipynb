```python
import pandas as pd

# Load the data from the CSV file
df = pd.read_csv(r'C:\Users\Mahesh.Dowlapall\Music\New folder\fictional_richest_people.csv')
df
```

```python
def convert_net_worth(net_worth):
    if 'billion' in net_worth:
        return float(net_worth.replace('$', '').replace(' billion', '')) * 1e9
    elif 'million' in net_worth:
        return float(net_worth.replace('$', '').replace(' million', '')) * 1e6
    else:
        return float(net_worth.replace('$', ''))
```

```python
df['Net Worth'] = df['Net Worth'].apply(convert_net_worth)
df
```

```python
# 1. Who is the richest in this list?
richest_person = df[df['Net Worth'] == df['Net Worth'].max()]['Name'].values[0]
print(f"The richest person in this list is: {richest_person}")

# 2. How many people are without an email?
people_without_email = df['Email'].isna().sum()
print(f"Number of people without an email: {people_without_email}")

# 3. How many people are without phone numbers?
people_without_phone = df['Phone Number'].isna().sum()
print(f"Number of people without a phone number: {people_without_phone}")
```
